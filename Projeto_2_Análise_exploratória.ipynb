## Qual segmento teve maior total de vendas?

## Qual total de vendas por segmento e por ano?

```python
# ## Qual o total de vendas por segmento e por ano?
# %%
# Converter a coluna 'Data_Pedido' para o tipo datetime, especificando o formato
df['Data_Pedido'] = pd.to_datetime(df['Data_Pedido'], format='%d/%m/%Y')

# Extrair o ano da data do pedido e criar uma nova coluna 'Ano'
df['Ano'] = df['Data_Pedido'].dt.year

# Agrupar por ano e segmento e somar o valor de venda
vendas_por_ano_segmento = df.groupby(['Ano', 'Segmento'])['Valor_Venda'].sum().reset_index()

# Para visualizar o resultado
print("Total de vendas por ano e segmento:")
print(vendas_por_ano_segmento)

# %%
import matplotlib.pyplot as plt
import seaborn as sns

# Criar o grÃ¡fico de linhas
plt.figure(figsize=(12, 7))
sns.lineplot(data=vendas_por_ano_segmento, x='Ano', y='Valor_Venda', hue='Segmento', marker='o')
plt.title('Total de Vendas por Ano e Segmento')
plt.xlabel('Ano')
plt.ylabel('Total de Vendas')
plt.xticks(vendas_por_ano_segmento['Ano'].unique()) # Define os ticks do eixo x nos anos Ãºnicos
plt.grid(True)
plt.legend(title='Segmento')
plt.show()
```

## Quantas vendas receberiam 15% de desconto?

```python
# ## Quantas vendas receberiam 15% de desconto (HipÃ³tese)?
# %%
# **HipÃ³tese:** Criando uma condiÃ§Ã£o de exemplo: Vendas com Valor_Venda acima de 1000 receberiam 15% de desconto.
# Substitua esta lÃ³gica pela sua condiÃ§Ã£o real de desconto.
desconto_condicao = df['Valor_Venda'] > 1000

# Criar uma nova coluna booleana indicando se a venda receberia desconto
df['Receberia_Desconto_15pct'] = desconto_condicao

# Contar o nÃºmero de vendas que receberiam o desconto
total_vendas_desconto = df['Receberia_Desconto_15pct'].sum()

print(f"NÃºmero de vendas que receberiam 15% de desconto (sob a hipÃ³tese): {total_vendas_desconto}")

# %% [markdown]
# **GrÃ¡fico:** Para visualizar a proporÃ§Ã£o de vendas com e sem desconto, um grÃ¡fico de pizza ou um countplot sÃ£o adequados. Vamos usar um countplot para mostrar as contagens.
# %%
import matplotlib.pyplot as plt
import seaborn as sns

# Criar um countplot para mostrar a contagem de vendas com/sem desconto
plt.figure(figsize=(8, 5))
sns.countplot(x='Receberia_Desconto_15pct', data=df, hue='Receberia_Desconto_15pct', palette='viridis', legend=False) # CorreÃ§Ã£o para evitar FutureWarning
plt.title('DistribuiÃ§Ã£o de Vendas com e sem Desconto (HipÃ³tese)')
plt.xlabel('Receberia 15% de Desconto')
plt.ylabel('NÃºmero de Vendas')
plt.xticks([False, True], ['NÃ£o', 'Sim']) # Define os rÃ³tulos do eixo x
plt.show()
```

## Qual o media de vendas por segmento, por ano e por mes?

```python
# ## Qual a mÃ©dia de vendas por segmento, por ano e por mÃªs?
# %%
# Converter a coluna 'Data_Pedido' para o tipo datetime (se ainda nÃ£o foi feito e corrigido o formato)
# Assumindo que o formato jÃ¡ foi corrigido para '%d/%m/%Y'
df['Data_Pedido'] = pd.to_datetime(df['Data_Pedido'], format='%d/%m/%Y')

# Extrair o ano e o mÃªs da data do pedido e criar novas colunas
df['Ano'] = df['Data_Pedido'].dt.year
df['Mes'] = df['Data_Pedido'].dt.month

# Agrupar por ano, mÃªs e segmento e calcular a mÃ©dia do valor de venda
media_vendas_por_periodo_segmento = df.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda'].mean().reset_index()

# Para visualizar o resultado
print("MÃ©dia de vendas por ano, mÃªs e segmento:")
print(media_vendas_por_periodo_segmento.head())

# %%
import matplotlib.pyplot as plt
import seaborn as sns

# Combinar Ano e MÃªs para um eixo x mais claro no grÃ¡fico
media_vendas_por_periodo_segmento['Ano_Mes'] = media_vendas_por_periodo_segmento['Ano'].astype(str) + '-' + media_vendas_por_periodo_segmento['Mes'].astype(str).str.zfill(2)

# Ordenar pelo Ano_Mes para garantir a ordem cronolÃ³gica no grÃ¡fico
media_vendas_por_periodo_segmento = media_vendas_por_periodo_segmento.sort_values(by='Ano_Mes')

# Criar o grÃ¡fico de linhas
plt.figure(figsize=(15, 8))
sns.lineplot(data=media_vendas_por_periodo_segmento, x='Ano_Mes', y='Valor_Venda', hue='Segmento', marker='o', palette='viridis') # 'viridis' Ã© uma paleta com cores vibrantes
plt.title('MÃ©dia de Vendas por PerÃ­odo (Ano-MÃªs) e Segmento')
plt.xlabel('PerÃ­odo (Ano-MÃªs)')
plt.ylabel('MÃ©dia do Valor de Venda')
plt.xticks(rotation=45, ha='right') # RotaÃ§Ã£o para melhor legibilidade dos rÃ³tulos de Ano-MÃªs
plt.grid(True)
plt.legend(title='Segmento')
plt.tight_layout() # Ajusta o layout
plt.show()
```
